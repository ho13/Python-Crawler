## 抓多頁的資料

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

table = []
for i in range(1,11):  #抓1-10頁的資料，每頁50筆
    url = 'http://ditu.92cha.com/dizhen.php?page={}&dizhen_ly=usa&dizhen_zjs=1&dizhen_zje=10&dizhen_riqis=1999-07-11&dizhen_riqie=2019-07-11'.format(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser') #獲取網頁html數據

    #抓資料
    row = []
    for tr in soup.find_all('tr'):
        #print(td.text.strip())
        line = []
        for trs in tr.find_all('td'):
            line.append(trs.text.strip())
        if line != [] : row.append(line)
    table.extend(row)  #每一頁的資料會存成list，用extend把每個list合併在一個list中，參考https://blog.csdn.net/xp5xp6/article/details/53635583
            
#抓欄位
col = []
for th in soup.find_all('th'):
    col.append(th.text.strip())

data = pd.DataFrame(table)
data.columns = col
data
```

## 只抓第一頁資料

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

url = 'http://ditu.92cha.com/dizhen.php?page=1&dizhen_ly=usa&dizhen_zjs=1&dizhen_zje=10&dizhen_riqis=1999-07-11&dizhen_riqie=2019-07-11'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser') #獲取網頁html數據

#抓資料
row = []
for tr in soup.find_all('tr'):
    #print(td.text.strip())
    line = []
    for trs in tr.find_all('td'):
        line.append(trs.text.strip())
    if line != [] : row.append(line)

#抓欄位
col = []
for th in soup.find_all('th'):
    col.append(th.text.strip())

data = pd.DataFrame(row)
data.columns = col
data
```
