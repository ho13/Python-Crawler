```python
name = []
sentence = '今天早上柯文哲表示要組台灣民眾黨，台銘表示非常興奮，蔡英文的選票可能要被瓜分了'
result = monpa.pseg(sentence)
for t in result:
    if (t[1] == 'PER'): name.append(t[0])
print('人名:', name)
    
```
