```python
soup = BeautifulSoup(resp.text, 'html.parser')
soup
```

```python
for tr in soup.find('table').find_all('tr'):
    print(tr.text.strip())
```

```python
cardinfo = {}
for tr in soup.find('table').find_all('tr'):
    th = tr.find('th').text
    th = th[:th.index(':')]
    td = tr.find('td').text
    cardinfo[th] = td
cardinfo
```
