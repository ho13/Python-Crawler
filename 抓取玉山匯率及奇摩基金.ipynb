## Part1:抓取玉山匯率

```python
#完整程式碼
import requests
from bs4 import BeautifulSoup

url = 'https://www.esunbank.com.tw/bank/personal/deposit/rate/forex/foreign-exchange-rates'
resp = requests.get(url)

soup = BeautifulSoup(resp.text, 'html.parser')
flag = 0
colnames = []
rowdata = []
for tr in soup.find_all('tr', {'class': 'tableContent-light'}):
    trs = tr.find_all('td')
    line = []
    for i in range(len(trs)):
        if (flag == 0): colnames.append(trs[i]['data-name'])
        line.append(trs[i].text.strip()) 
    flag = 1
    rowdata.append(line)
    
import pandas as pd
data =pd.DataFrame(rowdata, columns = colnames)
data
```

## Part2:抓取基金價格

```python
#完整程式碼
import requests
from bs4 import BeautifulSoup

url = 'https://tw.money.yahoo.com/fund/history/F0GBR04MS1:FO'
resp = requests.get(url)
resp.encoding = 'utf-8'  #搜尋charset可找到網頁編碼

soup = BeautifulSoup(resp.text, 'html.parser')

#抓資料值
row = []
for tr in soup.find_all('tr', {'class': 'H-36'}):
    line = []
    for trs in tr.find_all('td'):
        line.append(trs.text.strip())
    line.remove('') #若沒加這行的話會跑出空値
    row.append(line)      

#抓欄位名稱
head =[]
for th in soup.find_all('th', {'class': {'Py-8 Pend-12 Ta-end Fz-s Fw-n', 'Py-8 Ta-start Pstart-12 Pend-20 Fz-s Fw-n'}}):  
    head.append(th.text.strip())
head[0:4]   

import pandas as pd
data =pd.DataFrame(row)
data.columns = head[0:4]
data
```
